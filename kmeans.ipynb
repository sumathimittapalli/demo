```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

x = np.array([
    [1,2], [1,3], [1,4], [1,0],
    [10,2], [10,1], [10,3], [10,4],
    [5,2], [5,1], [5,3], [5,0],
    [8,2], [8,1], [8,3], [8,0]
])

kmeans = KMeans(n_clusters=2, random_state=0)
labels = kmeans.fit_predict(x)

print("Cluster Labels:", labels)
print("Centroids:\n", kmeans.cluster_centers_)

plt.scatter(x[:, 0], x[:, 1], c=labels, s=100)
plt.scatter(
    kmeans.cluster_centers_[:, 0], 
    kmeans.cluster_centers_[:, 1],
    s=300, marker='X', c='red'
)
plt.title("K-Means Clustering")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# -----------------------------
# Bigger Dataset (3 clusters)
# -----------------------------
x = np.array([
    [1,1], [1,2], [2,1], [2,2], [1,3], [2,3], [3,1], [3,2], [3,3], [0,1],
    [10,10], [10,11], [11,10], [11,11], [10,12], [12,10], [12,11], [11,12], [12,12], [9,10],
    [5,5], [5,6], [6,5], [6,6], [5,4], [4,5], [4,4], [6,4], [4,6], [7,5]
])

# -----------------------------
# Apply K-Means (3 clusters)
# -----------------------------
kmeans = KMeans(n_clusters=3, random_state=0)
labels = kmeans.fit_predict(x)

# Print results
print("Cluster Labels:", labels)
print("\nCentroids:\n", kmeans.cluster_centers_)

# -----------------------------
# Visualization
# -----------------------------
plt.scatter(x[:, 0], x[:, 1], c=labels, s=100, cmap='viridis')
plt.scatter(
    kmeans.cluster_centers_[:, 0], 
    kmeans.cluster_centers_[:, 1],
    s=300, marker='X', c='red'
)
plt.title("K-Means Clustering (3 Clusters)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

X = np.array([
    [1,1],[1,2],[2,1],[2,2],[1,3],[2,3],[3,1],[3,2],[3,3],[0,1],
    [10,10],[10,11],[11,10],[11,11],[10,12],[12,10],[12,11],[11,12],[12,12],[9,10],
    [5,5],[5,6],[6,5],[6,6],[5,4],[4,5],[4,4],[6,4],[4,6],[7,5]
])

agglo = AgglomerativeClustering(n_clusters=3)
labels = agglo.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels, s=100, cmap='viridis')
plt.title("Hierarchical Clustering (Agglomerative)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

X = np.array([
    [1,1],[1,2],[2,1],[2,2],[1,3],[2,3],[3,1],[3,2],[3,3],[0,1],
    [10,10],[10,11],[11,10],[11,11],[10,12],[12,10],[12,11],[11,12],[12,12],[9,10],
    [5,5],[5,6],[6,5],[6,6],[5,4],[4,5],[4,4],[6,4],[4,6],[7,5]
])

dbscan = DBSCAN(eps=2, min_samples=3)
labels = dbscan.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels, s=100, cmap='viridis')
plt.title("DBSCAN Clustering")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# Dataset: 4 clusters, with some noise points
X = np.array([
    [1,1],[2,1],[1,2],[2,2],
    [10,10],[11,10],[10,11],[11,11],
    [5,5],[6,5],[5,6],[6,6],
    [20,0],[21,1]   # noise points
])

dbscan = DBSCAN(eps=1.5, min_samples=2)
labels = dbscan.fit_predict(X)

plt.scatter(X[:,0], X[:,1],c=labels, s=100, cmap='viridis')
plt.title("DBSCAN Clustering (noise points included)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift

# Dataset: 3 clusters, non-uniform sizes
X = np.array([
    [1,1],[2,1],[1,2],[2,2],[1,3],
    [10,10],[11,10],[10,11],[11,11],[10,12],[12,10],
    [5,5],[6,5],[5,6],[6,6],[5,4],[4,5]
])

mean_shift = MeanShift()
labels = mean_shift.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels, s=100, cmap='viridis')
plt.scatter(mean_shift.cluster_centers_[:,0], mean_shift.cluster_centers_[:,1],
            s=300, marker='o', c='red')
plt.title("Mean-Shift Clustering (3 clusters)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Dataset: 3 clusters
X = np.array([
    [1,1],[2,1],[1,2],[2,2],
    [10,10],[11,10],[10,11],[11,11],
    [5,5],[6,5],[5,6],[6,6],
    [7,1],[8,1],[7,2],[8,2]
])

gmm = GaussianMixture(n_components=3, random_state=0)
labels = gmm.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels, s=100, cmap='viridis')
plt.title("GMM Clustering (3 clusters)")
plt.show()
```

```python
#dimensionality reduction
# PCA
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

np.random.seed(0)
x=np.random.rand(100, 3)

pca=PCA(n_components=2)
x_reduced=pca.fit_transform(x)

print("original shape:",x.shape)
print("reduced shape:",x_reduced.shape)
print("explained variance:",pca.explained_variance_ratio_)

plt.scatter(x_reduced[:,0],x_reduced[:,1])
plt.title("PCA - 3D to 2D Reduction")
plt.xlabel("pc1")
plt.ylabel("pc2")
plt.show()


```

```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
import matplotlib.pyplot as plt

# New dataset (different from previous one)
dataset = [
    ['eggs', 'milk', 'bread'],
    ['milk', 'diapers', 'beer', 'chips'],
    ['bread', 'butter'],
    ['eggs', 'bread', 'butter'],
    ['milk', 'diapers', 'bread'],
    ['beer', 'chips'],
    ['eggs', 'milk'],
]

# Convert to one-hot encoded format
te = TransactionEncoder()
te_data = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_data, columns=te.columns_)

# Apply Apriori
frequent_items = apriori(df, min_support=0.3, use_colnames=True)
print("Frequent Itemsets:\n", frequent_items)

# Create Association Rules
rules = association_rules(frequent_items, metric="confidence", min_threshold=0.5)
print("\nAssociation Rules:\n", rules)

# ============================
#  VISUALIZATION
# ============================

# 1 Support Chart
plt.figure(figsize=(6,4))
plt.bar(frequent_items['itemsets'].astype(str), frequent_items['support'])
plt.xlabel("Itemsets")
plt.ylabel("Support")
plt.title("Support of Frequent Itemsets")
plt.xticks(rotation=45)
plt.show()
```
